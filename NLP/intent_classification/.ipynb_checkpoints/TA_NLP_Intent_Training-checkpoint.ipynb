```python
# Step 2: Import Libraries
import pandas as pd
from datasets import Dataset
from transformers import BertTokenizer, BertTokenizerFast, BertForSequenceClassification, TrainingArguments, Trainer,  DataCollatorWithPadding
import torch
```

```python
#from google.colab import drive
#drive.mount('/content/drive')
#%cd /content/drive/MyDrive/SMU_MITB_NLP/Project/
```

```python
# Step 3: Load and Prepare Dataset
df = pd.read_csv("intent_classification_template.csv")  # Adjust path if needed

# Encode label column
label2id = {label: i for i, label in enumerate(df['intent'].unique())}
df['label'] = df['intent'].map(label2id)

max_len = 0
for t in df['text']:
  if len(t) > max_len:
    max_len = len(t.split(" "))

print(max_len)
```

```python
# Tokenizer
tokenizer = BertTokenizerFast.from_pretrained("bert-base-uncased")

def tokenize(batch):
    return tokenizer(batch["text"], padding=True, truncation=True)

# Convert to Hugging Face dataset
dataset = Dataset.from_pandas(df[["text", "label"]])
dataset = dataset.map(tokenize, batched=True)
dataset = dataset.train_test_split(test_size=0.2)

print(dataset)
```

```python
# View the first tokenized example in the training set

for sample in dataset['train']:
    print("="*60)
    print("Original Text:", sample['text'])
    print("Input IDs:", sample['input_ids'])
    print("Tokens:", tokenizer.convert_ids_to_tokens(sample['input_ids']))
    print("Token Type ID:", sample['token_type_ids'])
    print("Attention Mask:", sample['attention_mask'])
```

```python
# Step 4: Load Model
model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=len(label2id))

# Step 5: Training Configuration
training_args = TrainingArguments(
    output_dir="./intent_model",
    eval_strategy="epoch",
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=10,
    logging_dir="./logs",
    save_strategy="epoch",
    load_best_model_at_end=True,
)

# Step 6: Define Trainer
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset["train"],
    eval_dataset=dataset["test"],
    data_collator=data_collator
)

# Step 7: Train!
trainer.train()
```

```python
# Explicitly save the model after training completes
trainer.save_model("./intent_model")
```

```python
# Load trained model and tokenizer
model_path = "./intent_model"  # Change to your model directory
model = BertForSequenceClassification.from_pretrained(model_path)
tokenizer = BertTokenizerFast.from_pretrained(model_path)

# Put model in eval mode
model.eval()
```

```python
label2id = {'FindPlace': 0, 'BookFlight': 1, 'AskOpeningHour': 2, 'SearchHotel': 3, 'PlanItinerary': 4}
id2label = {v: k for k, v in label2id.items()}
```

```python
def predict_intent(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
    pred_id = torch.argmax(outputs.logits, dim=1).item()
    return id2label[pred_id]

# Example
test_query = "Can you suggest a hotel near Marina Bay?"
predicted_intent = predict_intent(test_query)
print(f" Predicted Intent: {predicted_intent}")
```

```python
test_queries = [
    "I'd like to visit Sentosa next week",
    "What's the best way to get to the airport?",
    "Book me a flight to Tokyo on Friday",
    "Where can I stay in Orchard?",
    "Plan a 2-day trip in Kyoto"
]

for q in test_queries:
    print(f"Query: {q}\n  Predicted Intent: {predict_intent(q)}\n")
```
